```python
import json
import requests

class Pokemon:
    list_of_pokemon = []
    
    def __init__(self, name=None, abilities=None, weight=None, type_of=None):
        self.name = name
        self.abilities = abilities
        self.weight = weight
        self.type_of = type_of
    
    def get_name(self):
        return self.name
    
    def get_abilities(self):
        return self.abilities
    
    def get_weight(self):
        return self.weight
    
    def get_type(self):
        return self.type_of
    
    def __repr__(self):
        return f'<Pokemon: {self.name} and {self.type_of} type>'
    
    def __str__(self):
        return f'{self.name} and {self.type_of} type'
    
#     @classmethod
    def all(self):
        print("+="*20)
        print(f"Pokemon Name: {self.name.title()}")
        print(f"Pokemon Ability: {self.abilities.title()}")
        print(f"Pokemon Weight: {self.weight}")
        print(f"Pokemon Type: {self.type_of.title()}")
        print("+="*20)
        print("\n")



class Pokedex:
    def __init__(self, name=None):
        self.web_link = f"https://pokeapi.co/api/v2/pokemon/{name}"
     
    def run(self):
        url = requests.get(self.web_link)
        if url.status_code == 400:
            print('You must executive a valid Pokemon name to execute this program.')
        elif url.status_code == 200:
            data = url.json()
            
            name = data['forms'][0]['name']
            abilities = data['abilities'][0]['ability']['name']
            weight = data['weight']
            type_of = data['types'][0]['type']['name']   
            
            p = Pokemon(name, abilities, weight, type_of)
            p.all()
    
    
# program = Pokedex("ditto")
# program.run()


list_of_pokemon = [
'rattata', 'pikachu', 'charmander', 'magikarp', 'ditto', 'snorlax', 'eevee', 'dragonite', 'koffing', 'weezing',
'moltres', 'kadabra', 'articuno', 'gastly', 'ekans', 'meowth', 'lucario', 'bulbasaur', 'articuno', 'mew'
    ]

for x in list_of_pokemon:
    program = Pokedex(x)
    program.run()

```
